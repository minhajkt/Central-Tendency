```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Dell/Downloads/house_price.csv")
df
```

```python
df.describe()
```

```python
#1 Mean Function
mean = df.price_per_sqft.mean()
mean
```

```python
#2 Percentile 
q1 = df['price_per_sqft'].quantile(0.25)
print(q1)
q2 = df['price_per_sqft'].quantile(0.50)
print(q2)
q3 = df['price_per_sqft'].quantile(0.75)
print(q3)
```

```python
#3 IQR
iqr = q3 - q1
print(iqr)
```

```python
# Finding Outliers
lower_whisker = q1-1.5*iqr
lower_whisker
```

```python
upper_whisker = q3+1.5*iqr
upper_whisker
```

```python
outliers = df[(df.price_per_sqft < -308) | (df.price_per_sqft > 11892)]
outliers
```

```python
outliers.describe()
```

```python
non_outliers = df[(df.price_per_sqft > -308) * (df.price_per_sqft < 11892)]
non_outliers
```

```python
df["price_per_sqft"].describe().round()
```

```python
#4 Normal Distribution
sns.histplot(data = df, x = df.price_per_sqft, kde=True)
plt.show()
```

```python
#5 Z Score
#zscore = (x - m)/s
m = df.price_per_sqft.mean()
s = df.price_per_sqft.std()
df['Zscore'] = (df.price_per_sqft - m) / s
df
```

```python
sns.boxplot(data = df)
plt.show()
```
